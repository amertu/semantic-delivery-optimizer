```python
names_store_ids = []


for sid in store_ids:
    store = data_vs['storesMap'][sid]
    
    uuids.append(store['uuid'])  # for identification
    
    title = store['title']
    title_only = title.split('(')[0]
    
    names_store_ids.append((title_only, store['uuid']))
    
    names.append(title_only)
    
    delivery = store['fareBadge']
    if delivery is not None:
        fare = re.findall("\d+\.\d+", delivery['text'])
        deliv_charge.append(float(fare[0]))
    else:
        deliv_charge.append(np.nan)

    price_cat.append(store['categories'][0] if not None else np.nan)

    cuisines.append(store['categories'][1:])
    try:
        rating.append(store['feedback']['rating'])
    except TypeError:
        rating.append(np.nan)
    try:
        rating_count.append(store['feedback']['ratingCount'])
    except TypeError:
        rating_count.append(np.nan)
    
    try:
        deliv_time_str = store['etaRange']['text']  # [:4]
        numbers_chars = [c for c in deliv_time_str if c.isnumeric()]
    
        if len(numbers_chars) == 2:
            # could be wrong for e.g. 5-7 minutes
            dur = float(''.join(numbers_chars))
        elif len(numbers_chars) == 3:
            dur = np.mean([float(''.join(numbers_chars[0])), float(''.join(numbers_chars[1:]))])
        elif len(numbers_chars) == 4:
            dur = np.mean([float(''.join(numbers_chars[:2])), float(''.join(numbers_chars[2:]))])
        else:
            dur = np.nan

        delivery_dur.append(dur)
    except TypeError:
        # print( store['etaRange'])  # this is typically None
        delivery_dur.append(np.nan)
    
    
```

```python
# collect menus for the restaurants, save to folder,
# one file per get_menu_responsesget_menu_responses = False

def strip_name(name_to_strip):
    """
    """
    return name_to_strip.strip().replace('\'', '').replace('-', '').replace(' ', '_').replace('/', '')


get_menu_responses = False

if get_menu_responses:
    menu_data = np.empty((len(names_store_ids), ), dtype=object)
    rest_w_menus_names = []

    for kk, name_store_id in enumerate(names_store_ids):
#         name = name_store_id[0].strip().replace('\'', '').replace('-', '').replace(' ', '_').replace('/', '')
        name = strip_name(name_store_id[0])
        store_id = name_store_id[1]

        # print(name)
        # print(store_id)
        rest_w_menus_names.append(name)

        # set up store uuid
        data = '{' + '"storeUuid":"{0}"'.format(store_id)  + ', "sfNuggetCount":1}'

        # get the data
        response_menu = requests.post('https://www.ubereats.com/api/getStoreV1', headers=headers, params=params, data=data)  # , cookies=cookies)

        try:
            resp_json = json.loads(response_menu.content)
            menu_data = resp_json['data']
        except ValueError:  # JSONDecodeError:
            print('Error, problem with response data')
            print(response_menu.content)
            continue  # skip this entry...

        # save menu items to a dedicated file
        with open(os.path.join(folder_details, name + '_' + store_id[:4] + '_details.json'), 'w') as jf:
            json.dump(menu_data, jf)
            jf.close()
```

```python
# list of details files
detail_files = os.listdir(folder_details)
```

```python
# get data from dumps of menu for restaurants

menu_data = {}
restaurant_locations = {}
restaurant_addresses = {}
restaurant_uuid = {}

for df in detail_files:
    print(df)  # print file name
    
    if not df.endswith('.json'):  # only use json file
        continue
        
    with open(os.path.join(folder_details, df), 'r') as jf:
        menu = json.load(jf)
        jf.close()

    try:
        loc_data = menu['location']  # dict with location data
        latitude = loc_data['latitude']
        longitude = loc_data['longitude']
        post_code = loc_data['postalCode']
        street_address = loc_data['streetAddress']
        city = loc_data['city']
        
        if street_address == post_code:
            street_address = ''
            
        leading_nr = re.findall('^\d+\s[A-z]+\s[A-z]+', street_address)
        leading_nr_2 = re.findall('^\d+\s[A-z]+', street_address)
        if leading_nr or leading_nr_2:
            number = street_address.split(' ')[0].strip()
            street_address = ' '.join(street_address.split(' ')[1:]).strip()
        else:
            number = None
        
        print('address: ', street_address)
        print('number: ', number)
        print('city: ', city)
        print(latitude, longitude)

    except KeyError:
        print('no location data')

    try:
        menues = menu['sectionEntitiesMap']
        keys = list(menues.keys())
        
        menu_items_list = []
       
        for key in keys:
            menu_items_keys = menues[key]
            
            for mi_key in menu_items_keys:  # iterate over the menu items
                # print(mi_key)
                menu_item = menues[key][mi_key]
                price_pounds =  str(menu_item['price'] / 100)
                name = menu_item['title']
                description = menu_item['description'] or ''
                
                menu_items_list.append([name, description, price_pounds])

            print('------> ' + 'no of menu items {}'.format(len(menu_items_list)))

        menu_data[df] = menu_items_list        
        restaurant_locations[df] = [longitude, latitude]
        restaurant_uuid[df] = menu['uuid']
        
        restaurant_addresses[df] = {'street': street_address,
                                    'post_code': post_code,
                                    'city': city,
                                    'number': number}
            
    except KeyError:
        print('no setctionEntitiesMap in data')
        print('\n')
        # continue
```
