```python
def load_src(fn):
    """
    """
    path_fn = os.path.join(data_folder, fn)
    
    with open(path_fn, 'r') as jf:
        data = json.load(jf)
        jf.close()

    return data
```

```python
# uber eats
df_uber_eats = pd.read_json(os.path.join(data_folder, 'uber_eats_with_menu.json'))
```

```python
# foodhub
df_foodhub = pd.read_json(os.path.join(data_folder, 'foodhub_with_menu.json'))
```

```python
df_supper = pd.read_json(os.path.join(data_folder, 'supper_london_with_menu.json'))
```

```python
# df_rest_concat = pd.concat([df_deliveroo, df_uber_eats, df_foodhub]).reset_index(drop=True)
df_rest_concat = pd.concat([df_uber_eats, df_foodhub, df_supper]).reset_index(drop=True)
```

```python
df_rest_concat.head(2)
```

```python
# clean up data
#   remove entries having no street
df_rest_concat = df_rest_concat.loc[df_rest_concat['street'] != '']
df_rest_concat = df_rest_concat.loc[pd.notnull(df_rest_concat['street'])]

#   remove entries having no price
df_rest_concat = df_rest_concat.loc[pd.notnull(df_rest_concat['price'])]

#   remove entries having no rating for the restaurant
df_rest_concat = df_rest_concat.loc[pd.notnull(df_rest_concat['rating_value'])]
```

```python
# store the dataframe with the concatenated restaurants df's
df_rest_concat.to_json(os.path.join(data_folder, 'merged_data.json'))
```
